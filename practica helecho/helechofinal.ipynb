```python
import random
from numpy import *
n=int(input("ingresa n "))
def chances():
    random_number=random.randint(1,100)
    
    if random_number<=1:
        return [0,0,0,0.16,0,0]
    elif 2<=random_number<=86:
        return [0.85,0.04,-0.04,0.85,0.,1.6]
    elif 87<=random_number<=93:
        return [0.2,-0.26,0.23,0.22,0.,1.6]
    elif 94<=random_number<=100:
        return [-0.15,0.28,0.26,0.24,0.,0.44]

lista1=chances()
lista2=[0,0]
x1=[]
x2=[]
def function(l,x):
    #A=matrix([[l[0],l[1]],[l[2],l[3]]])
    vector=matrix([[l[4]],[l[5]]])

    i=0
    while i<n:
        B=matrix([[l[0]*x[0]+l[1]*x[1]],[l[2]*x[0]+l[3]*x[1]]])
        F=B+vector
        x[0]=F[0,0]
        x[1]=F[1,0]
        l=chances()
        x1.append(x[0])
        x2.append(x[1])
        i+=1
    
function(lista1,lista2)
import matplotlib.pyplot as plt
plt.scatter(x1, x2, s = 0.2, edgecolor ='green')

```
